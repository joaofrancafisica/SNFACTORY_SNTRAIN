```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Read The data

```python
train_data = pd.read_csv("list.dat",
                         header=None,
                         sep="\s+",
                         names=["name", "phase", "file_name"])
```

## Number of spectra distribution

```python
plt.hist(train_data["name"].value_counts())
plt.xlabel("$N_{spec}$")
plt.ylabel("N");
```

## Number of phase distribution

```python
plt.hist(train_data["phase"], bins=50)
plt.xlabel("$Time (days)$")
plt.xlim(-20, 50)
plt.ylabel("N");
```
